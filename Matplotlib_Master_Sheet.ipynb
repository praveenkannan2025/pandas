Once installed, you can import Matplotlib into your Python script or Jupyter notebook. The most commonly used submodule for plotting is pyplot, which provides a MATLAB-like interface.

````python
import matplotlib.pyplot as plt
````

<font color='teal'>Example Code</font>

````python
import matplotlib.pyplot as plt

# Sample data
x = [1, 2, 3, 4, 5]
y = [2, 3, 5, 7, 11]

# Create a figure and axis
fig, ax = plt.subplots()

# Plot data
ax.plot(x, y, label='Line Plot')

# Add a title and labels
ax.set_title('Simple Line Plot')
ax.set_xlabel('X-axis')
ax.set_ylabel('Y-axis')

# Add a legend
ax.legend()

# Display the plot
plt.show()
````

- <font color='navy'>Plotting the Data</font>: `ax.plot()` is used to plot the data on the axes.
- <font color='navy'>Customizing the Plot</font>: We set the title, x-label, and y-label.
- <font color='navy'>Adding a Legend</font>: `ax.legend()` adds a legend to the plot.
- <font color='navy'>Displaying the Plot</font>: `plt.show()` renders the plot in a window.


# <font color='blue'>2. Basic Plotting with Matplotlib</font>

Matplotlib makes it easy to create a variety of plots, with line plots being one of the most fundamental types. In this section, we’ll cover how to create line plots, customize their appearance, and add labels to enhance readability.

## <font color='darkgreen'>Line Plots</font>

Line plots are useful for visualizing data points connected by lines. They are often used to display trends over time or continuous data.

### <font color='teal'>Creating Line Plots</font>

To create a basic line plot, you use the `plot()` function from Matplotlib’s `pyplot` module. Here’s a simple example:

```python
import matplotlib.pyplot as plt

# Sample data
x = [1, 2, 3, 4, 5]
y = [2, 3, 5, 7, 11]

# Create a line plot
plt.plot(x, y)

# Display the plot
plt.show()


- In this example, `plt.plot(x, y)` generates a line plot with the data provided.

<font color='teal'>Customizing Line Style</font>

Matplotlib allows you to customize the appearance of lines in your plot. You can modify colors, markers, and line styles using parameters in the plot() function.

- <font color='navy'>Colors</font>: You can specify the color of the line using standard color names or hexadecimal color codes.
- <font color='navy'>Markers</font>: Markers can be used to highlight data points.
- <font color='navy'>Line Styles</font>: You can choose from various line styles like solid, dashed, and dotted.

````python
import matplotlib.pyplot as plt

# Sample data
x = [1, 2, 3, 4, 5]
y = [2, 3, 5, 7, 11]

# Create a line plot with customization
plt.plot(x, y, color='green', linestyle='--', marker='o', label='Data Line')

# Add a legend
plt.legend()

# Display the plot
plt.show()
````

<font color='teal'>Adding Labels</font>

To make your plots more informative, you can add labels and titles. This includes setting the x-axis label, y-axis label, and plot title.

````python
import matplotlib.pyplot as plt

# Sample data
x = [1, 2, 3, 4, 5]
y = [2, 3, 5, 7, 11]

# Create a line plot
plt.plot(x, y, color='blue', linestyle='-', marker='x', label='Data Line')

# Add labels and title
plt.xlabel('X-axis Label')
plt.ylabel('Y-axis Label')
plt.title('Line Plot Example')

# Add a legend
plt.legend()

# Display the plot
plt.show()
````

# <font color='blue'>3. Subplots and Grids with Matplotlib</font>

Matplotlib allows for advanced plotting capabilities, including creating multiple plots within a single figure using subplots and grid specifications. This guide covers the basics of setting up subplots and adjusting their layout.

## <font color='darkgreen'>3. Subplots and Grids</font>

### <font color='teal'>Creating Multiple Plots</font>

Matplotlib provides functions to create multiple plots within a single figure, allowing for effective comparison and visualization of different datasets or aspects of the same dataset.

#### **Subplots in One Figure**

The `subplot()` function allows you to create multiple subplots within a single figure. Each subplot is specified by its position in a grid.

Here’s an example:

```python
import matplotlib.pyplot as plt

# Create a figure
plt.figure(figsize=(10, 6))

# First subplot
plt.subplot(2, 2, 1)
plt.plot([1, 2, 3], [4, 5, 6])
plt.title('Plot 1')

# Second subplot
plt.subplot(2, 2, 2)
plt.scatter([1, 2, 3], [4, 5, 6])
plt.title('Plot 2')

# Third subplot
plt.subplot(2, 2, 3)
plt.bar(['A', 'B', 'C'], [4, 5, 6])
plt.title('Plot 3')

# Fourth subplot
plt.subplot(2, 2, 4)
plt.hist([1, 2, 2, 3, 4])
plt.title('Plot 4')

# Display the figure
plt.tight_layout()  # Adjust layout
plt.show()


**In this example:**

- `plt.subplot(2, 2, 1)` creates a 2x2 grid and places the plot in the first position.
- Each subsequent call to `plt.subplot()` places additional plots in the specified grid positions.

````python
import matplotlib.pyplot as plt

# Create subplots
fig, axs = plt.subplots(2, 2, figsize=(10, 6))

# First subplot
axs[0, 0].plot([1, 2, 3], [4, 5, 6])
axs[0, 0].set_title('Plot 1')

# Second subplot
axs[0, 1].scatter([1, 2, 3], [4, 5, 6])
axs[0, 1].set_title('Plot 2')

# Third subplot
axs[1, 0].bar(['A', 'B', 'C'], [4, 5, 6])
axs[1, 0].set_title('Plot 3')

# Fourth subplot
axs[1, 1].hist([1, 2, 2, 3, 4])
axs[1, 1].set_title('Plot 4')

# Adjust layout
plt.tight_layout()
plt.show()
````

**In this example:**

- `fig, axs = plt.subplots(2, 2)` creates a 2x2 grid of subplots.
- `axs[i, j]` accesses the subplot in row i and column j.

<font color='teal'>Adjusting Layout</font>
- To ensure that subplots are displayed clearly without overlapping, Matplotlib provides methods for adjusting the layout.

**Tight Layout**
- The `tight_layout()` function automatically adjusts subplot parameters to give specified padding and prevent overlap.

````python
import matplotlib.pyplot as plt

# Create a figure with subplots
fig, axs = plt.subplots(2, 2, figsize=(10, 6))

# Add plots
axs[0, 0].plot([1, 2, 3], [4, 5, 6])
axs[0, 1].scatter([1, 2, 3], [4, 5, 6])
axs[1, 0].bar(['A', 'B', 'C'], [4, 5, 6])
axs[1, 1].hist([1, 2, 2, 3, 4])

# Automatically adjust layout
plt.tight_layout()
plt.show()
````

**Adding Spacing**
- The subplots_adjust() function allows for manual adjustment of the spacing between subplots.

````python
import matplotlib.pyplot as plt

# Create a figure with subplots
fig, axs = plt.subplots(2, 2, figsize=(10, 6))

# Add plots
axs[0, 0].plot([1, 2, 3], [4, 5, 6])
axs[0, 1].scatter([1, 2, 3], [4, 5, 6])
axs[1, 0].bar(['A', 'B', 'C'], [4, 5, 6])
axs[1, 1].hist([1, 2, 2, 3, 4])

# Manually adjust subplot parameters
plt.subplots_adjust(left=0.1, right=0.9, top=0.9, bottom=0.1, wspace=0.4, hspace=0.4)
plt.show()
````

**In this example:**

- `left`, `right`, `top`, `bottom` control the spacing around the edges of the figure.
- `wspace` and `hspace` control the spacing between subplots horizontally and vertically.

**e. Subplots with Grid Specification**

````python
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D
import numpy as np

# Create data
x = np.linspace(-5, 5, 100)
y = np.linspace(-5, 5, 100)
x, y = np.meshgrid(x, y)
z = np.sin(np.sqrt(x**2 + y**2))

# Create a figure
fig = plt.figure(figsize=(12, 8))

# Add 3D subplot
ax = fig.add_subplot(111, projection='3d')
ax.plot_surface(x, y, z, cmap='plasma', edgecolor='none')
ax.plot_wireframe(x, y, z, color='black')

# Add title and labels
ax.set_title('Customized 3D Surface Plot', fontsize=16, fontstyle='italic', color='darkblue')
ax.set_xlabel('X axis')
ax.set_ylabel('Y axis')
ax.set_zlabel('Z axis')

# Show the plot
plt.show()
````

#### <font color='teal'>Adding Interactive Widgets</font>
- Interactive widgets can enhance your plots by adding elements like sliders, buttons, and dropdowns to adjust parameters dynamically.

**Interactive Plots in Jupyter Notebooks**
- You can use `mpl_toolkits.mplot3d` and `ipywidgets` to add interactivity. Here’s an example using sliders to control the parameters of a plot.

````python
import numpy as np
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D
from ipywidgets import interact

# Function to update the plot
def update_plot(freq=1):
    x = np.linspace(-5, 5, 100)
    y = np.linspace(-5, 5, 100)
    x, y = np.meshgrid(x, y)
    z = np.sin(freq * np.sqrt(x**2 + y**2))
    
    fig = plt.figure(figsize=(12, 8))
    ax = fig.add_subplot(111, projection='3d')
    ax.plot_surface(x, y, z, cmap='viridis')
    
    ax.set_title('Interactive 3D Plot', fontsize=16, fontstyle='italic', color='darkblue')
    ax.set_xlabel('X axis')
    ax.set_ylabel('Y axis')
    ax.set_zlabel('Z axis')
    
    plt.show()

# Create an interactive widget
interact(update_plot, freq=(1, 10, 1))
````

- you can use a slider to adjust the frequency of the sine wave in a 3D plot. The `interact` function from `ipywidgets` makes it easy to create interactive widgets.

**Using Widgets with Plots**
- Widgets like sliders, buttons, and dropdowns allow for more complex interactions with your plots. Here’s an example using sliders and dropdowns.

````python
import matplotlib.pyplot as plt
from ipywidgets import interact, widgets
import numpy as np

# Function to update the plot
def plot_function(amplitude=1.0, frequency=1.0):
    x = np.linspace(0, 10, 100)
    y = amplitude * np.sin(frequency * x)
    
    plt.figure(figsize=(10, 6))
    plt.plot(x, y)
    plt.title(f'Sine Wave: Amplitude={amplitude}, Frequency={frequency}', fontsize=16, fontstyle='italic', color='darkblue')
    plt.xlabel('X Axis')
    plt.ylabel('Y Axis')
    plt.grid(True)
    plt.show()

# Create interactive widgets
interact(plot_function, 
         amplitude=widgets.FloatSlider(value=1.0, min=0, max=2, step=0.1, description='Amplitude'),
         frequency=widgets.FloatSlider(value=1.0, min=0, max=10, step=0.1, description='Frequency'))
````

- sliders are used to control the amplitude and frequency of a sine wave. The plot updates automatically as you adjust the sliders.
